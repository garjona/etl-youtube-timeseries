## 3. Conexión a la Base de Datos PostgreSQL y carga de los datos en DataFrames  

En esta sección, se establece la conexión con la base de datos PostgreSQL para recuperar los datos almacenados de la API de YouTube. Los datos serán utilizados para realizar el análisis.  

Pasos:  
1. Se utiliza un archivo `.env` para almacenar las credenciales de la base de datos de forma segura.  
2. Con **SQLAlchemy**, se configura un motor de conexión a PostgreSQL y se recuperan los datos desde las tablas relevantes:  
   - **video_metadata**: Contiene metadatos de los videos, como título, duración y fecha de publicación.  
   - **youtube_video_data**: Almacena datos temporales de las métricas de visualización de los videos.  
3. Los datos se cargan en DataFrames de pandas para facilitar su análisis posterior.  


```python
# Especificar la ruta del archivo .env
dotenv_path = os.path.join(os.path.dirname(os.path.dirname(os.path.abspath('Notebook'))), '.env')
print("Ruta del archivo .env:", dotenv_path)

# Cargar variables de entorno desde .env
config = dotenv_values(dotenv_path)

# Parámetros de conexión a PostgreSQL
host = config["DB_HOST"]
database = config["DB_DATABASE"]
user = config["DB_USER"]
password = config["DB_PASSWORD"]
port = config["DB_PORT"]

# Crear el engine de SQLAlchemy para PostgreSQL
engine = create_engine(f'postgresql://{user}:{password}@{host}:{port}/{database}')

# Crear una sesión para interactuar con la base de datos
Session = sessionmaker(bind=engine)
session = Session()

# Consultas SQL para obtener los datos de interés
query_video_metadata = """
SELECT video_id, title, duration, published_at
FROM video_metadata;
"""

query_youtube_video_data = """
SELECT video_id, views, timestamp, metadata_version
FROM youtube_video_data
ORDER BY timestamp;
"""

# Leer los datos en DataFrames utilizando el motor SQLAlchemy
df_metadata = pd.read_sql(query_video_metadata, engine)
df_video_data = pd.read_sql(query_youtube_video_data, engine)

# Mostrar un resumen inicial de los datos obtenidos
print("Metadatos de los videos:")
print(df_metadata.head())

print("\nDatos temporales de YouTube:")
print(df_video_data.head())

# Cerrar la sesión para liberar recursos
session.close()

# Liberar el motor de conexión
engine.dispose()
```

### 6.1 Análisis: Evolución Temporal de las Visualizaciones Totales

En esta sección, analizamos la evolución de las **visualizaciones totales** a lo largo del tiempo, específicamente por minuto. Este análisis es importante para entender cómo varían las visualizaciones en un periodo de tiempo concreto, en este caso, limitado a un solo día con registros por minuto.

#### Objetivo:
Observar el comportamiento de las visualizaciones a nivel temporal, para identificar patrones o momentos clave en los que las visualizaciones tienden a aumentar o disminuir.

#### Pasos realizados:
1. **Selección de videos consistentes**:  
   - Se identificaron los `video_id` que están presentes en todos los minutos del período de análisis. Solo los videos que tienen registros en **todos los minutos** se consideran en este análisis.
   
2. **Agrupación de datos por minuto**:  
   - Las visualizaciones fueron agrupadas por minuto, sumando las visualizaciones de los videos que cumplen con el criterio anterior.

3. **Visualización**:  
   - Se creó un gráfico de líneas para mostrar la evolución de las visualizaciones totales a lo largo del tiempo, específicamente para los videos que estuvieron presentes en todos los minutos.

Este análisis nos permitirá entender mejor la dinámica temporal de las visualizaciones y sentar las bases para un análisis más detallado en futuras etapas.



```python
# Asegurarse de que la columna 'timestamp' esté en formato datetime
df_merged['timestamp'] = pd.to_datetime(df_merged['timestamp'])

# Identificar los 'video_id' presentes en todos los minutos
# Contamos cuántos minutos están disponibles en total
total_minutes = len(df_merged['timestamp'].unique())

# Identificamos los 'video_id' que están presentes en todos los minutos
video_ids_in_all_minutes = df_merged.groupby('video_id')['timestamp'].nunique()
video_ids_in_all_minutes = video_ids_in_all_minutes[video_ids_in_all_minutes == total_minutes].index

# Filtrar los datos solo con los 'video_id' presentes en todos los minutos
df_filtered = df_merged[df_merged['video_id'].isin(video_ids_in_all_minutes)]

# Agrupar por 'timestamp' y 'video_id' para obtener las visualizaciones por minuto para cada video
views_by_minute_video = df_filtered.groupby(['video_id', 'timestamp'])['views'].sum().reset_index()

# Agrupar por 'timestamp' (todos los videos) para obtener las visualizaciones totales por minuto
views_by_minute_total = views_by_minute_video.groupby('timestamp')['views'].sum().reset_index()

# Crear el gráfico de evolución de las visualizaciones totales por minuto
plt.figure(figsize=(12, 6))
plt.plot(views_by_minute_total['timestamp'], views_by_minute_total['views'], color='dodgerblue', marker='o', linestyle='-', linewidth=2)

# Etiquetas y título
plt.title('Evolución de las Visualizaciones Totales por Minuto (solo videos_id consistentes)', fontsize=14)
plt.xlabel('Hora:Minuto', fontsize=12)
plt.ylabel('Total de Visualizaciones', fontsize=12)

# Mejorar la visualización de las fechas y horas en el eje X (mostrar hora y minuto)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ajustar el diseño
plt.tight_layout()

# Mostrar el gráfico
plt.show()

```

### 6.2. Comparación de la Evolución de Visualizaciones en Videos Recientes vs. Antiguos

En este análisis, se comparará la evolución de las visualizaciones de los videos en función de su fecha de publicación, diferenciando entre videos más recientes y videos más antiguos. El objetivo es identificar si los videos más recientes tienen una tendencia diferente en cuanto a visualizaciones a lo largo del tiempo.

#### Objetivo:
Comparar la tasa de crecimiento de las visualizaciones entre videos publicados en diferentes fechas para ver si los videos recientes tienen una mayor o menor tasa de crecimiento en sus visualizaciones.

#### Pasos realizados
1. **Agrupación por día de publicación**: Los videos se agruparon en función del día de su publicación.
2. **Cálculo de la tasa de crecimiento**: La tasa de crecimiento se calcula como el cambio en el número de visualizaciones cada 3 minutos.
3. **Gráfico de Líneas**: Se generará un gráfico de líneas, donde cada línea representará la evolución de las visualizaciones para videos publicados en un día específico, permitiendo comparar cómo los videos más recientes han evolucionado en términos de visualizaciones frente a los más antiguos.

Este análisis proporcionará una visión más detallada del comportamiento temporal de las visualizaciones de los videos, centrándose en la tasa de crecimiento, lo que puede ofrecer indicios sobre las preferencias de la audiencia o el impacto de los videos más recientes.





```python
# Asegurarse de que 'timestamp' y 'published_at' estén en formato datetime
df_merged['timestamp'] = pd.to_datetime(df_merged['timestamp'])
df_merged['published_at'] = pd.to_datetime(df_merged['published_at'])

# Extraer solo la fecha de la publicación (sin la hora) para poder agrupar por días
df_merged['publish_day'] = df_merged['published_at'].dt.date

# Identificar los 'video_id' que están presentes en todos los minutos
total_minutes = len(df_merged['timestamp'].unique())
video_ids_in_all_minutes = df_merged.groupby('video_id')['timestamp'].nunique()
video_ids_in_all_minutes = video_ids_in_all_minutes[video_ids_in_all_minutes == total_minutes].index

# Filtrar los datos para los videos consistentes
df_filtered = df_merged[df_merged['video_id'].isin(video_ids_in_all_minutes)]

# Agrupar por 'publish_day', 'timestamp' y 'video_id' para obtener las visualizaciones por minuto
views_by_minute_video_grouped = df_filtered.groupby(['publish_day', 'timestamp', 'video_id'])['views'].sum().reset_index()

# Agrupar por 'publish_day' y 'timestamp' para obtener las visualizaciones totales por minuto
views_by_minute_grouped_total = views_by_minute_video_grouped.groupby(['publish_day', 'timestamp'])['views'].sum().reset_index()

# Agrupar por intervalos de 3 minutos en el eje X
views_by_minute_grouped_total['timestamp_3min'] = views_by_minute_grouped_total['timestamp'].dt.floor('3T')  # Redondear al intervalo de 3 minutos

# Tomar solo el último registro de cada intervalo de 3 minutos
views_by_minute_grouped_total_last = views_by_minute_grouped_total.groupby(['publish_day', 'timestamp_3min']).last().reset_index()

# Ahora vamos a calcular la tasa de crecimiento entre los intervalos de 3 minutos
# Calculamos la tasa de crecimiento de las visualizaciones (porcentaje de cambio entre los intervalos de 3 minutos)
views_by_minute_grouped_total_last['growth_rate'] = views_by_minute_grouped_total_last.groupby(['publish_day'])['views'].pct_change().fillna(0) * 100  # Tasa de crecimiento en porcentaje

# Filtramos los datos para quedarnos solo con los intervalos de 3 minutos
views_by_minute_grouped_total_last = views_by_minute_grouped_total_last.dropna(subset=['growth_rate'])

# Crear el gráfico de comparación
plt.figure(figsize=(12, 6))

# Graficar las líneas para cada día de publicación
for day in views_by_minute_grouped_total_last['publish_day'].unique():
    data_day = views_by_minute_grouped_total_last[views_by_minute_grouped_total_last['publish_day'] == day]
    plt.plot(data_day['timestamp_3min'], data_day['growth_rate'], label=f'Publicación {day}', marker='o', linestyle='-', linewidth=2)

# Etiquetas y título
plt.title('Comparación de la Tasa de Crecimiento de Visualizaciones en Videos Publicados en Diferentes Días (Cada 3 Minutos)', fontsize=14)
plt.xlabel('Hora:Minuto (Intervalo de 3 minutos)', fontsize=12)
plt.ylabel('Tasa de Crecimiento (%)', fontsize=12)

# Mejorar la visualización de las fechas y horas en el eje X
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Leyenda
plt.legend()

# Ajustar el diseño
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

#### Análisis 6.2: Tasa de Crecimiento de Visualizaciones en Videos Publicados en Diferentes Días (Cada 3 Minutos)

En este análisis, se evaluó la tasa de crecimiento de las visualizaciones de videos publicados en diferentes días. La tasa de crecimiento fue calculada para intervalos de 3 minutos, tomando los datos de visualizaciones a cada minuto y luego calculando el porcentaje de cambio entre los intervalos de 3 minutos.

#### Conclusiones análisis:
- Se puede observar que, en general, la tasa de crecimiento es baja en la mayoría de los casos, con valores cercanos a 0% en varios momentos.
- Algunos días muestran picos de crecimiento en ciertos intervalos de 3 minutos, como el 2024-12-09 y 2024-12-10, donde las tasas de crecimiento son mayores.
- Los videos publicados más recientemente (en el 2024-12-09 y 2024-12-10) tienden a tener tasas de crecimiento más altas en comparación con los publicados en días anteriores.
